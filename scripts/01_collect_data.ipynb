# Data from Gsponer 2016

this script parses data from Gsponer training and testing datasets

## Proteins from train dataset

```python
# CRE-containing proteins in training dataset (Gsponer 2016)
# It contains both disordered and structured CREs
cre_train = pd.read_excel('../raw_data/1-s2.0-S2405471216300345-mmc2.xlsx', skiprows= 3)
```

```python
cre_train.columns = cre_train.columns.str.lower()
cre_train.rename(columns= {'uniprot accession number': 'uniprot_acc'}, inplace= True)
```

```python
cre_train.info()
```

```python
# Drop bibliography and empty rows
cre_train = cre_train.dropna()
cre_train
```

```python
# Index of the position where the other table beggins
ix = cre_train.index[cre_train.name == 'Name'].to_list()[0]
```

```python
# df of Proteins with a disordered CREs
cre_train_disordered = cre_train.loc[:ix-1]
cre_train_disordered['module'] = 'd'
```

```python
# df of proteins with structured CREs
#  ix is the header
cre_train_structured = cre_train.loc[ix+1:]
cre_train_structured['module'] = 's'
```

```python
cre_train = pd.concat([cre_train_disordered, cre_train_structured])
cre_train
```

```python
# Select only the uniprot accesion
cre_train.uniprot_acc = cre_train.uniprot_acc.apply(lambda x: x.split(':')[1])
```

```python
cre_train.species.value_counts()
```

```python
# strip blank spaces
cre_train.species = cre_train.species.str.strip()
```

```python
cre_train.species.value_counts()
```

see

```python
# One protein could have one or more CREs, separated by ","
cre_train.cre = cre_train.cre.str.split(",")
cre_train = cre_train.explode("cre")
```

```python
# Format start and end aa
cre_train['start_aa'] = cre_train.cre.apply(lambda x: x.split('-')[0]).astype(int)
cre_train['end_aa'] = cre_train.cre.apply(lambda x: x.split('-')[1]).astype(int)
cre_train.drop(columns= 'cre', inplace= True)
```

```python
cre_train.info()
```

## Proteins from test dataset
to do: format correctly start and end

```python
cre_test = pd.read_excel('../raw_data/1-s2.0-S2405471216300345-mmc3.xlsx', skiprows= 3)
```

```python
# Format cols
cre_test.columns = cre_test.columns.str.lower()
cre_test.rename(columns= {'uniprot accession number': 'uniprot_acc', 'crea': 'cre'}, inplace= True)
```

```python
# Drop references
ix = cre_test.index[cre_test.name == 'References'].to_list()[0]
cre_test = cre_test.loc[:ix-1]
# Drop empty rows
cre_test = cre_test.dropna()
```

```python
cre_test
```

```python
ix = cre_test.index[cre_test.name == 'Name'].to_list()[0]
cre_test_disordered = cre_test.loc[:ix-1]
cre_test_disordered['module'] = 'd'
cre_test_structured = cre_test.loc[ix+1:]
cre_test_structured['module'] = 's'
```

```python
cre_test = pd.concat([cre_test_disordered, cre_test_structured])
```

```python
# strip blanks
cre_test.species = cre_test.species.str.strip()
```

```python
cre_test.species.value_counts()
```

```python
# Select only the uniprot accesion
cre_test.uniprot_acc = cre_test.uniprot_acc.apply(lambda x: x.split(':')[1])
```

```python
# Some proteins may have more than one CRE module
cre_test.cre = cre_test.cre.str.split(',')
cre_test = cre_test.explode('cre')
```

```python
# Format start and end aa
cre_test['start_aa'] = cre_test.cre.apply(lambda x: x.split('-')[0]).astype(int)
cre_test['end_aa'] = cre_test.cre.apply(lambda x: x.split('-')[1]).astype(int)
cre_test.drop(columns= 'cre', inplace= True)
```
