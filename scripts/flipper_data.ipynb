## FLIPPER data
an automatic method for the detection of structurally linear sub-regions or peptides that interact with another chain in a protein complex.

### Json file

```python
# Regex for uniprot accesions
pattern = re.compile('[OPQ][0-9][A-Z0-9]{3}[0-9]|[A-NR-Z][0-9]([A-Z][A-Z0-9]{2}[0-9]){1,2}') # uniprot accession patterns
```

```python
# File to write the matches
to_add = open("../raw_data/filtered_flipper.js", "a+")
# Iterate the FLIPPER json file
with open("../raw_data/flipper_mobidb_result_2022-07-07T14_09_17.692Z.json", "r") as f:
    i = 0
    for line in f:
        try:
            acc = re.search(pattern, line).group(0)
            if acc in up:
                i += 1
                to_add.write(line)
        except:
            pass
to_add.close()
print(f'found {i} proteins')
```

# Check overlap with CRE elements

```python
# df with the ranges for each CRE. Rename cols for pyranges
cre_ranges = cre_dataset[['uniprot_acc', 'id_cre', 'start_aa', 'end_aa']].rename(
    columns={'uniprot_acc': 'Chromosome', 'start_aa': 'Start', 'end_aa': 'End'}
)
# As pyranges object
cre_ranges = pr.PyRanges(cre_ranges)
```

```python
cre_ranges
```

```python
# LIPs with their ranges
lip_ranges = fl_consensus[['uniprot_acc', 'start', 'end']].rename(
    columns = {'uniprot_acc': 'Chromosome'}
)
lip_ranges.columns = lip_ranges.columns.str.capitalize()
# As pyranges object
lip_ranges = pr.PyRanges(lip_ranges)
```

```python
lip_ranges
```

```python
# Count overlaps
overlap = cre_ranges.count_overlaps(lip_ranges, overlap_col = 'overlap_count')
```

```python
# to dataframe
overlap_df = overlap.as_df()
overlap_df.sort_values("overlap_count")
```

```python
xx = overlap[overlap.overlap_count > 0]
xx
```
